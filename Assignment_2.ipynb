### 6. Create a list:
```python
numbers = [12, 45, 7, 23, 56, 89, 34]

Write a program to find:

* Largest element
* Second largest element
* Smallest element
```

````python
numbers=[12,45,7,23,56,89,34]

Large=numbers[0]
Small=numbers[0]

# Taking Big Negative Similar to INT_MIN
S_Large=-50000

for i in numbers:
    if(i>Large):
        Large=i
    if(i<Small):
        Small=i
for i in numbers:
    if(i>S_Large and i!=Large):
        S_Large=i

print("Largest:",Large)
print("Smallest:",Small)
print("Second Largest:",S_Large)
````

### 7. Create a list:
```python
arr = [10, 20, 30, 40, 50, 60]

Write a function that takes the list as input and returns the list in reverse order without using the `reverse()` method.
```

````python
def Rev(lst):
    start=0
    end=len(lst)-1
    while(start<end):
        temp=lst[start]
        lst[start]=lst[end]
        lst[end]=temp
        start=start+1
        end=end-1
    return lst

arr=[10,20,30,40,50,60]
rev=Rev(arr)
print("Array after Reverse:",rev)
````

### 10. Write a function:
```python
count_frequency(arr)

that takes a list as input and prints the frequency of each element.

Example:

arr = [1, 2, 2, 3, 1, 4, 2]
```

Output:

```text
1 -> 2 times
2 -> 3 times
3 -> 1 time
4 -> 1 time
```

````python
def count_freq(arr):
    Freq={}
    for i in arr:
        key=Freq.keys()
        if(i not in key):
            Freq[i]=1
        else:
            Freq[i]=Freq[i]+1
    for key,value in Freq.items():
        print(key,"->",value,"times")
arr=[1,2,2,3,1,4,2]
count_freq(arr)
````
