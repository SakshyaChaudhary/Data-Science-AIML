### 1. Write a program that takes a list of numbers and prints all prime numbers in the list using loops.

### Do not use any built-in prime-checking functions.
```

````python
def Prime(num):
    isPrime=True
    for i in range(2,int(num**0.5)+1):
        if (num%i==0):
            isPrime=False
            break
    if (isPrime==True and num>1):
        print(num)
Lst=[10, 7, 15, 23, 8, 31, 42, 17, 9, 29]
for i in Lst:
    Prime(i)
````

### 3. Write a user-defined function that accepts a list of integers and returns:
- Count of even numbers
- Sum of odd numbers

Use loops and conditional statements inside the function.
```

````python
def Love(Lst):
    count=0
    summ=0
    for i in Lst:
        if(i%2==0):
            count+=1
        else:
            summ+=i
    return count,summ
Lst=[1,2,3,4,5,6,7]
nEven,sumOdd=Love(Lst)
print("Count of Even:",nEven)
print("Sum of Odd:",sumOdd)
````

### 4. Create a function that calculates simple interest.

Use default arguments for rate and time.

Demonstrate calling the function using both positional and keyword arguments.
```


````python
# SI = (Principal × Rate × Time) / 100
# Principal (P) = Amount borrowed or invested
# Rate (R) = Rate of interest per year (%)
# Time (T) = Time in years
def SimpleInterest(Principal,Rate=10,Time=2):
    SI=(Principal*Rate*Time)/100
    return SI
print("Default Argument:")
Amount=5000
print("Principal is",Amount,"and SI is",SimpleInterest(Amount))
print()

print("Positional Argument with Rate=8% and Time=4 years:")
print("Principal is",Amount,"and SI is",SimpleInterest(Amount,8,4))
print()

print("Keyword Argument with Rate=12% and Time=5 years:")
print("Principal is",Amount,"and SI is",SimpleInterest(Principal=Amount,Time=5,Rate=12))

````

### 5. Create a Student class with attributes:

### 7. Create a BankAccount class with private attributes for balance.

Provide methods to deposit and withdraw money.

Ensure balance cannot be accessed directly.
```

````python
class BankAccount:
    def __init__(self,name,balance):
        self.name=name
        self.__balance=balance
    def Deposit(self,amount):
        self.__balance=self.__balance+amount
    def Withdraw(self,amount):
        self.__balance=self.__balance-amount
    
    def Display(self):
        print("Account Balance:",self.__balance)

ob=BankAccount("Sakshya Chaudhary",5000)
ob.Display()
print("Deposit:500")
ob.Deposit(500)
ob.Display()
print("Withdraw:300")
ob.Withdraw(300)
ob.Display()
print()

print("Account Holder:",ob.name)
# print("Balance using private:",ob.__balance)
# This doesnt work in Python because balance is private
````
