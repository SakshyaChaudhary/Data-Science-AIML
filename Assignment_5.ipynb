### 1. Create a 1D NumPy array of 20 random integers between 1 and 100. Find the minimum, maximum, mean, and standard deviation without using Python built-in functions — use only NumPy aggregation functions.

### 6. Create a DataFrame with some missing values intentionally placed in the Marks and City columns. Then:
- Identify missing values using isnull()
- Fill missing marks with the column mean using fillna()
- Drop rows where City is missing using dropna()

```python
data = {
    "Name": ["Sakshya", "Anurag", "Harsh", "Aman", "Priya"],
    "Age": [20, 19, 21, 18, 20],
    "Marks": [85, np.nan, 91, np.nan, 77],
    "City": ["Delhi", "Mumbai", np.nan, "Jaipur", np.nan]
}

df = pd.DataFrame(data)

print("\nMissing Values:")
print(df.isnull())

df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

df.drop("City",axis=1,inplace=True)

print("\nAfter Dropping Rows with Missing City:")
print(df)
```

### 9. Using Matplotlib, create the following plots from a student marks dataset (create the dataset yourself):
- A line plot showing marks trend across 5 subjects
- A bar chart comparing marks of 5 students
- Add proper title, xlabel, ylabel, legend, and grid to both plots

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Name": ["Sakshya", "Anurag", "Harsh", "Aman", "Priya"],
    "Maths": [85, 72, 91, 68, 77],
    "Science": [88, 75, 89, 70, 80],
    "English": [82, 78, 85, 65, 79],
    "Computer": [95, 80, 93, 72, 88],
    "Hindi": [79, 70, 84, 67, 75]
}

df = pd.DataFrame(data)

subjects = ["Maths", "Science", "English", "Computer", "Hindi"]
marks = [85, 88, 82, 95, 79]  # Sakshya's marks

plt.figure(figsize=(8, 5))
plt.plot(subjects, marks,
         marker="o",
         color="blue",
         linewidth=2,
         label="Sakshya")

plt.title("Marks Trend Across 5 Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.legend()
plt.grid()
plt.show()
students = df["Name"]
total_marks = df[["Maths", "Science", "English", "Computer", "Hindi"]].sum(axis=1)

plt.figure(figsize=(8, 5))
plt.bar(students, total_marks,
        color="green",
        label="Total Marks")

plt.title("Comparison of Total Marks of 5 Students")
plt.xlabel("Students")
plt.ylabel("Total Marks")
plt.legend()
plt.grid()
plt.show()
```

### 10. Using Pandas built-in plotting, load or create a dataset with at least 4 numeric columns. Create:
- A histogram for marks distribution
- A box plot to visualize spread and outliers
- A pie chart showing category-wise percentage

```python
data = {
    "Maths": [85, 72, 91, 68, 77, 88, 95, 74, 81, 79],
    "Science": [88, 75, 89, 70, 80, 85, 92, 78, 83, 76],
    "English": [82, 78, 85, 65, 79, 84, 90, 72, 80, 75],
    "Computer": [95, 80, 93, 72, 88, 91, 97, 76, 85, 82],
    "Category": ["A", "B", "A", "C", "B", "A", "A", "C", "B", "B"]
}

df = pd.DataFrame(data)

# Histogram
df["Maths"].plot(
    kind="hist",
    bins=5,
    title="Maths Marks Distribution"
)

plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Box Plot
df[["Maths", "Science", "English", "Computer"]].plot(
    kind="box",
    title="Box Plot of Student Marks"
)

plt.ylabel("Marks")
plt.show()

# Pie Chart
df["Category"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    title="Category-wise Percentage"
)

plt.ylabel("")
plt.show()
```

### 11. Using Seaborn, create a dataset of at least 50 records with columns Age, Salary, Department, Experience. Plot:
- A scatter plot of Age vs Salary with hue as Department
- A heatmap showing correlation between numeric columns
- A bar plot showing average Salary per Departmen

```python
import seaborn as sns

np.random.seed(42)

df = pd.DataFrame({
    "Age": np.random.randint(22, 50, 50),
    "Salary": np.random.randint(25000, 100000, 50),
    "Department": np.random.choice(
        ["HR", "IT", "Finance", "Marketing"], 50
    ),
    "Experience": np.random.randint(1, 20, 50)
})

print(df.head())

plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x="Age",
    y="Salary",
    hue="Department"
)

plt.title("Age vs Salary by Department")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()

plt.show()

plt.figure(figsize=(6, 4))

corr = df[["Age", "Salary", "Experience"]].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()

plt.figure(figsize=(8, 5))

sns.barplot(
    data=df,
    x="Department",
    y="Salary"
)

plt.title("Average Salary per Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.grid()

plt.show()
```

### 12. Create a subplot layout of 2 rows × 2 columns using Matplotlib containing:
- Line plot
- Scatter plot
- Bar chart
- Histogram
All using the same dataset. Add a main title using suptitle()

```python
data = {
    "Name": ["Sakshya", "Anurag", "Harsh", "Aman", "Priya"],
    "Marks": [85, 72, 91, 68, 77],
    "Age": [20, 19, 21, 18, 20]
}
df = pd.DataFrame(data)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axes[0, 0].plot(df["Name"], df["Marks"],
                marker="o",
                label="Marks")
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("Students")
axes[0, 0].set_ylabel("Marks")
axes[0, 0].legend()
axes[0, 0].grid()

# Scatter Plot
axes[0, 1].scatter(df["Age"], df["Marks"],
                   marker="o",
                   label="Students")
axes[0, 1].set_title("Scatter Plot")
axes[0, 1].set_xlabel("Age")
axes[0, 1].set_ylabel("Marks")
axes[0, 1].legend()
axes[0, 1].grid()

# Bar Chart
axes[1, 0].bar(df["Name"], df["Marks"],
               label="Marks")
axes[1, 0].set_title("Bar Chart")
axes[1, 0].set_xlabel("Students")
axes[1, 0].set_ylabel("Marks")
axes[1, 0].legend()
axes[1, 0].grid()

# Histogram
axes[1, 1].hist(df["Marks"],
                bins=5,
                label="Marks")
axes[1, 1].set_title("Histogram")
axes[1, 1].set_xlabel("Marks")
axes[1, 1].set_ylabel("Frequency")
axes[1, 1].legend()
axes[1, 1].grid()

# Main Title
fig.suptitle("Student Dataset Visualization", fontsize=16)

plt.tight_layout()
plt.show()
```
